```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Timestamp
import matplotlib.dates as mdates
```

```python
uri='https://raw.githubusercontent.com/PatapongPanpiboon/thermospheric-density/main/Sunsport_24to25.csv'
df=pd.read_csv(uri)
df.Date = pd.to_datetime(df.Date)
df.set_index('Date', inplace=True)
df.head()
```

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots(1, 1, figsize=(12, 4))

# Creating the plot
ax.scatter(df.index, df['Daily total sunspot number'],s = 5, color ='tab:blue')
d_monthly = df['Daily total sunspot number'].rolling(30, center = True).mean()
d_yearly = df['Daily total sunspot number'].rolling(365, center = True).mean()
ax.plot(d_monthly,lw = 2, color = 'darkorange')
ax.plot(d_yearly,lw = 2, color = 'tab:red')
ax.axvspan(xmin = '2018-05-29', xmax = '2022-12-25', facecolor="steelblue", edgecolor="none", alpha=0.5)
ax.set_ylim(0,250)
ax.set_xlabel('Date [UT]', size=16)
ax.set_ylabel("Sunspot number", size=16);
ax.set_title('Sunspot number data of WDC-SILSO from January 1, 2009, to August 31, 2023', size=16)
ax.legend(['Daily', 'Monthly', 'Yearly', 'GRACE-FO'],markerscale=2 ,edgecolor='white',loc = 'upper left')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.savefig('SunspotNumber_WholeDataset.png', dpi=240)
plt.margins(0, 0.1)
plt.show()
```
